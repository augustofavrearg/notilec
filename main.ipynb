```python
import requests
response = requests.get("https://www.infobae.com")
print("La resopnse es: ", dir(response))
print(response.text)
```

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')
notes_links = soup.select('.story-card-img-ctn')
notes = [note['href'] for note in notes_links]

for note in notes:
    print(note)
```
